## **과제 1**

**API**
- API란 "Application Programming Interface"의 줄임말로, 두 소프트웨어 간의 의사소통을 가능하게 만들어주는 통신 메커니즘이다. 여기서 "Application"은 특정 기능을 갖춘 소프트웨어를, "Interface"는 소프트웨어끼리 요청 및 응답을 이용해 서로 정보를 주고 받는 방식을 정의한 계약이다. 예를 들어 핸드폰 앱으로 내 금융 정보를 보고자 한다면, 카카오뱅크의 소프트웨어와 핸드폰 앱이 정해진 인터페이스에 따라 정보를 교환하여 나의 계좌 잔액을 보여주게 된다.

**REST API**
- REST는 특정 리소스를 URI를 명시하고 HTTP 방식을 이용해 해당 리소스에 대한 Create, Read, Update, 그리고 Delete 연산을 적용하는 것을 의미한다. 이 원리를 따르는 API를 REST API라고 부르고, 기존 HTTP 방식을 그대로 가져다 쓰기 때문에 새로운 시스템을 구축할 필요가 없을뿐더러 명확한 메시지로 정확한 커뮤니케이션을 가능하게 한다.

**Create**
- GET

**Read**
- POST

**Update**
- PUT / PATCH

**Delete**
- DELETE

```python
import requests
import json
import time

url = 'https://api.punkapi.com/v2/beers?food=crab'
## 게와 잘어울리는 맥주가 먹고싶다.
headers = {'Content-Type': 'application/json'}
params= {'page': 1, 'per_page':10 }

response = requests.get(url, headers=headers,params=params)

if response.status_code == 200:
    result = response.json()
    print('받아온 데이터:', result)
else:
    print('요청에 실패했습니다. 상태 코드:', response.status_code)
```

```python
import pandas as pd

df = pd.DataFrame(result)
df
```
