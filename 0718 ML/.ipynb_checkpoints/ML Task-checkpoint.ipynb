```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_validate
from sklearn.model_selection import StratifiedKFold
from lightgbm import LGBMClassifier
from lightgbm import plot_importance
import matplotlib.pyplot as plt

print("------------------------- Data Preparation -------------------------")

## 와인 데이터 받기
wine = pd.read_csv('https://bit.ly/wine_csv_data')

## 넘피로 변환 후 훈련/테스트 스플릿
data = wine[['alcohol', 'sugar', 'pH']].to_numpy()
target = wine['class'].to_numpy()
train_input, test_input, train_target, test_target = train_test_split(data, target, test_size=0.2, random_state=42)

## 데이터 표준화 전처리
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)

print("------------------------- Model Training -------------------------")

## LGBMClassider 파라미터 딕셔너리 생성 후 모델 생성
params_wrapper = {
    'boosting_type': 'gbdt',  ## traditional 그래디언트 부스팅 결정트리 사용
    'num_leaves': 31,         ## 트리 리프 수 제한
    'max_depth': 1,           ## 트리 깊이 제한
    'learning_rate': 0.1,     ## 학습률 설정. 천천히 목표치로 다가감
    'n_estimators': 200,      ## 만들 트리 개수 지정
    'objective': 'binary',    ## 이진 분류가 목적임
    'min_child_samples': 20,  ## 리프 하나에 들어갈 데이터 개수 최소값 설정
    'bagging_fraction': 0.8,  ## 각 iteration당 사용할 데이터의 fraction.
    'n_jobs': -1              ## 스레드 전부 사용
}

## 모델 생성
lgbm_wrapper = LGBMClassifier(**params_wrapper)

## Cross Validation을 위한 타겟값 튜플의 리스트 생성
evals = [(test_input, test_target)]

## Fit 함수에 들어갈 파라미터 딕셔너리 생성
params_fit = {
    'eval_set': evals,
    'eval_metric': 'logloss',
    'verbose': True
}

## 모델에 트레이닝 데이터 적용
lgbm_wrapper.fit(train_input, train_target, **params_fit)

print("------------------------- Cross Validation -------------------------")

## 분류 모델에 사용되는 StratifiedKFold를 이용, 섞은 데이터로 CV 10번 반복.
splitter = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
scores = cross_validate(lgbm_wrapper, train_input, train_target, cv=splitter)

print("Train data 정확도 점수               : ",lgbm_wrapper.score(train_input,train_target))
print("Train data 교차 검증 정확도 평균 점수: ",np.mean(scores['test_score']))

print("------------------------- Test Data Accuracy -------------------------")

print("Test data 정확도 점수                : ",lgbm_wrapper.score(test_input,test_target))

print("------------------------- Feature Importance -------------------------")

fig, ax = plt.subplots(figsize=(10,5))
plot_importance(lgbm_wrapper,ax=ax)
ax.set_yticklabels(['Alcohol','Sugar','pH'])
plt.show()

```

## 과제 2: 그리드 서치 VS. 랜덤 서치

그리드 서치는 사용자가 직접 지정해준 하이퍼 파라미터를 이용해 loss 함수의 극소값을 찾지만, 최적해에 무조건적으로 도달한다는 보장이 없고, 서치 범위가 커질수록 연산에 오랜 시간이 걸립니다. 따라서 파라미터의 범위가 굉장히 좁은 케이스에서나 효과적이라 생각합니다. 반면에 랜덤 서치는 우리가 최적화 하고자하는 하이퍼 파라미터의 범위가 클수록 정해진 iteration 내에서 굉장히 높은 확률로 최적해를 찾을 수 있습니다. 또한, 그리드 서치가 격자 형태로 검색값들을 지정해준다면, 랜덤 서치는 말그대로 형식에 얽매이지 않고 랜덤하게 값들을 뽑아서 검사하기 때문에 더 다양한 조합들을 시도해볼 수 있다는 장점이 있습니다.

## 과제 3: Decision Tree regression task 진행해보기

```python
## Seaborn 내장 미국 환경보호국 MPG 데이터 활용
import seaborn as sns
df = sns.load_dataset("mpg")
```

1) cylinders, displacement, horsepower, weight, acceleration, model_year 으로 해당 자동차의 mpg를 regression 모델로 예측해보고 싶다.  
2) horsepower에 결측치 6개 있다. 몇개 안되니까 그냥 해당 행 삭제해버리자.  

```python
df = df.dropna(subset=['horsepower'],axis=0)
df = df.drop(columns=['name','origin'])
df.reset_index()
```

```python
df.columns
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import KFold
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform, randint


print("------------------------- Data Preparation -------------------------")

## 넘피로 변환 후 훈련/테스트 스플릿
data = df[['cylinders','displacement','horsepower','weight','acceleration','model_year']].to_numpy()
target = df['mpg'].to_numpy()
train_input, test_input, train_target, test_target = train_test_split(data, target, test_size=0.2, random_state=42)

## 데이터 표준화 전처리
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)

print("------------------------- Model Training -------------------------")

## 모델 생성
dtr = DecisionTreeRegressor()

## 모델에 트레이닝 데이터 적용
dtr.fit(train_input, train_target)

print("------------------------- Tuning & Cross Validation -------------------------")

## 랜덤 서치에 이용할 파라미터 범위
params = {
    'max_depth': randint(1,50),
    'min_samples_leaf': randint(1,10),
    'min_samples_split': randint(1,10),
    'max_leaf_nodes': randint(1,50) 
}

## 하이퍼 파라미터 튜닝
random_search = RandomizedSearchCV(dtr,params,random_state=42)
random_search.fit(train_input, train_target)
tuned_params = random_search.best_params_

## 최적 모델 받기
dtr_tuned = DecisionTreeRegressor(**tuned_params)

## 피팅
dtr_tuned.fit(train_input, train_target)

print("------------------------- Results -------------------------")

print("Train data 정확도 점수               : ",dtr_tuned.score(train_input,train_target))
print("Test data 정확도 점수                : ",dtr_tuned.score(test_input,test_target))
```
